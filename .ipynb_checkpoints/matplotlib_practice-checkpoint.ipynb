```python
# Create a Line Chart Using the MATLAB Method
# make sure our plots are displayed inline within the Jupyter Notebook directly below the code cell that produced it
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Create our Arrays or lists 

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create a Line Chart Using the Object-Oriented Interface

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Or fig, ax = plt.subplots() can be written as...
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

# Where fig references the figure, and ax references the axes of the figure and allows us to annotate the figure.
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
# To create a quick graph using the object-oriented interface method, use the following two lines:

# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# To change figure-level attributes (such as axis labels, a title, or a legend) or save the figure as an image, 
# use fig = plt.figure().
```

```python
# Use plt.show()

# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Add features to our graph 

# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create a Vertical Bar Chart
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create Scatter Plots and Bubble Charts
import matplotlib.pyplot as plt
```

```python
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]
```

```python
# Create a Scatter Plot Using the MATLAB Method: plt.plot()

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis, y_axis, 'o')
```

```python
# Create a scatter plot using the MATLAB Method: plt.scatter()
plt.scatter(x_axis, y_axis)
```

```python
#Skill Drill

# Using the Matplotlib MATLAB plotting approach, make the following changes to your scatter plot:
plt.scatter(y_axis, x_axis, c='red', marker='o', label='Chicago')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.legend()
plt.gca().invert_yaxis()
# Change the color of the markers to red.
# Add a legend for the city of Chicago.
# Add a title and axes labels.
# Switch the axis so that the Fare($) data is on the x-axis.
# Add a limit for the x-axis data.
# Invert the y-axis so that January is at the top.
```

```python
# Create a Bubble Chart Using the MATLAB Method
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Increase bubble size 
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# List comprehension instead of a for or while loop
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# SKILL DRILL
# Using the object-oriented approach, make the following changes to your bubble chart:

# Change the color of the markers to sky blue.
# Change the size of the markers to 5 times each data point.

# Make the color 20% transparent.
# Add a black edge color to the circles.
# Make the linewidth of the circles 2 points.

# Add a legend for the city of Boston.
# Add a title.
# Switch the axis so that the Fare($) data is on the x-axis.
# Add a limit for the x-axis data.
# Invert the y-axis so that January is at the top.

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], c='skyblue', marker='o', label='Boston')
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.legend()
plt.gca().invert_yaxis()
```

```python
# Create a Pie Chart Using the MATLAB Method
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Customize pie chart 
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Customize pie chart 
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface

fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```
